```python
hyd_dataset.drop_duplicates(inplace=True)
hyd_dataset.shape
```

```python
columns_with_outliers = []

outlier_counts = {}

for column in hyd_dataset.columns:
    if hyd_dataset[column].dtype != 'object':
        plt.figure(figsize=(6, 4))
        plt.boxplot(hyd_dataset[column], vert=False)
        plt.title(f'Box Plot of {column}')
        plt.xlabel(column)
        plt.show()

        Q1 = hyd_dataset[column].quantile(0.25)
        Q3 = hyd_dataset[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = hyd_dataset[(hyd_dataset[column] < lower_bound) | (hyd_dataset[column] > upper_bound)]
        if not outliers.empty:
            columns_with_outliers.append(column)
            outlier_counts[column] = len(outliers)
```

```python
sorted_outliers = dict(sorted(outlier_counts.items(), key=lambda x: x[1], reverse=True))

for column, count in sorted_outliers.items():
    print(f"Column: {column}, Outliers: {count}")
```

```python
outlier_features = ['Area','No. of Bedrooms']
for feature in outlier_features:
    IQR = hyd_dataset[feature].quantile(0.75) - hyd_dataset[feature].quantile(0.25)
    lower_boundary = hyd_dataset[feature].quantile(0.25) - (IQR*1.5)
    upper_boundary = hyd_dataset[feature].quantile(0.75) + (IQR*1.5)
    hyd_dataset.loc[hyd_dataset[feature]<=lower_boundary, feature] = lower_boundary
    hyd_dataset.loc[hyd_dataset[feature]>=upper_boundary, feature] = upper_boundary
```

```python
features = hyd_dataset.drop(['Price', 'Location'], axis=1)
label = hyd_dataset['Price']
features.shape, label.shape
```

```python
hyd_dataset = pd.get_dummies(hyd_dataset)
hyd_dataset.shape
```

```python
hyd_dataset.columns
```

```python
features = ['MaintenanceStaff','Gymnasium','SwimmingPool','LandscapedGardens','JoggingTrack','RainWaterHarvesting',
            'IndoorGames','ShoppingMall','Intercom','SportsFacility','ATM','ClubHouse','School','24X7Security',
            'PowerBackup','CarParking','StaffQuarter','Cafeteria','MultipurposeRoom','Hospital','WashingMachine',
            'AC','Wifi','BED','VaastuCompliant','Microwave','GolfCourse','TV','DiningTable',
            'Sofa','Refrigerator']
selected_features = hyd_dataset.drop(hyd_dataset[features], axis=1)
from sklearn.model_selection import train_test_split
features_train, features_test, label_train, label_test = train_test_split(selected_features, label, test_size=0.3, random_state=40)
features_train.shape, features_test.shape, label_train.shape, label_test.shape
```

```python
n_estimators = [int(x) for x in np.linspace(start=100, stop=1200, num=12)]

max_features = ['auto', 'sqrt']

max_depth = [int(x) for x in np.linspace(start=5, stop=30, num=6)]

min_samples_split = [2, 5, 10, 15, 100]

min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators' : n_estimators,
              'max_features' : max_features,
              'max_depth' : max_depth,
              'min_samples_split' : min_samples_split,
              'min_samples_leaf' : min_samples_leaf}
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
random_forest = RandomForestRegressor()
random_forest_model = RandomizedSearchCV(estimator=random_forest, param_distributions=random_grid, n_jobs=1, random_state=42,
                                        cv=5, n_iter=10, verbose=2, scoring='neg_mean_squared_error')
random_forest_model.fit(features_train, label_train)
```

```python
label_pred = random_forest_model.predict(features_test)
label_pred
```

```python
from sklearn.metrics import r2_score
r2_score(label_train, random_forest_model.predict(features_train))
```

```python
r2_score(label_test, random_forest_model.predict(features_test))
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(label_test, label_test, label='Actual Prices', alpha=0.5, marker='o', color='blue')
plt.scatter(label_test, label_pred, label='Predicted Prices', alpha=0.5, marker='x', color='red')
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs. Predicted Prices using Random Forest")
plt.legend()
plt.show()
```
